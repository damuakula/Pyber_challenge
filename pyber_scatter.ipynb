```python
%matplotlib inline

# Setup Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Setup City and Ride data from csv files
city_data_to_load = "C:\\Users\\Owner\\GitHub\\Pyber_challenge\\city_data.csv"
ride_data_to_load = "C:\\Users\\Owner\\GitHub\\Pyber_challenge\\ride_data.csv"

# Read the City and Ride Data from csv files
ride_data = pd.read_csv(ride_data_to_load)
city_data =  pd.read_csv(city_data_to_load)

# Combine the City and Ride data into a single dataset
pyber_df = pd.merge(ride_data, city_data, how='left', on='city')

# Display the combined data table for preview
pyber_df
```

```python
group_type_df = pyber_df.groupby(['type'])
group_type_df
```

```python
group_city_df = pyber_df.groupby(['city'])
group_city_df
```

```python
average_fare_per_type = group_type_df["fare"].mean()
average_fare_per_type
```

```python
average_fare_per_city = group_city_df["fare"].mean()
```

```python
total_rides_per_type = group_type_df["ride_id"].count()
total_rides_per_type
```

```python
total_rides_per_city = group_city_df["ride_id"].count()
```

```python
colors = np.random.rand(len(total_rides_per_city))
area = (30 * np.random.rand(len(total_rides_per_city)))**2

#fig, ax = plt.subplots()
#ax.scatter(total_rides_per_city, average_fare_per_city, s=area, c=colors, facecolors="red", edgecolors="black", alpha=0.5)
plt.scatter(total_rides_per_city, average_fare_per_city, s=area, c=colors, facecolors="red", edgecolors="black", alpha=0.5)

#ax.set_xlabel('Total Number of Rides (Per City)', fontsize=15)
#ax.set_ylabel('Average Fare($)', fontsize=15)
#ax.set_title('Pyber Ride Sharing Data (2016)',fontsize=18)

plt.xlabel('Total Number of Rides (Per City)', fontsize=15)
plt.ylabel('Average Fare($)', fontsize=15)
plt.title('Pyber Ride Sharing Data (2016)',fontsize=18)

plt.grid(True)
#fig.tight_layout()
#plt.legend()

plt.show
```
