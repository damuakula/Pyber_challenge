```python
%matplotlib inline

# Setup Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Setup City and Ride data from csv files
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data from csv files
ride_data = pd.read_csv(ride_data_to_load)
city_data =  pd.read_csv(city_data_to_load)

# Combine the City and Ride data into a single dataset
pyber_df = pd.merge(ride_data, city_data, how='left', on='city')

# Display the combined data table for preview
pyber_df
```

```python
group_city_df = pyber_df.groupby(['city'])
group_city_df
```

```python
group_type_df = pyber_df.groupby(['type'])
group_type_df["ride_id"].count()
```

```python
urban_city_df = pyber_df.loc[(pyber_df['type'] == "Urban")]
suburban_city_df = pyber_df.loc[(pyber_df['type'] == "Suburban")]
rural_city_df = pyber_df.loc[(pyber_df['type'] == "Rural")]
```

```python
type_fare_group = group_type_df["fare"].sum()
```

```python
type_driver_group = group_type_df["driver_count"].sum()
```

```python
percent_total_fares_urban = round((type_fare_group["Urban"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_fares_suburban = round((type_fare_group["Suburban"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_fares_rural = round((type_fare_group["Rural"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
```

```python
percent_total_rides_urban = round((len(urban_city_df)/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_rides_suburban = round((len(suburban_city_df)/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_rides_rural = round((len(rural_city_df)/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
```

```python
percent_total_drivers_urban = round((type_driver_group["Urban"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_drivers_suburban = round((type_driver_group["Suburban"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
percent_total_drivers_rural = round((type_driver_group["Rural"]/(len(urban_city_df) + len(suburban_city_df) + len(rural_city_df)))*100,2)
```

```python
urban_group = urban_city_df.groupby(['city'])
suburban_group = suburban_city_df.groupby(['city'])
rural_group = rural_city_df.groupby(['city'])
```

```python
driver_count_urban_city = urban_group["driver_count"].count()
driver_count_suburban_city = suburban_group["driver_count"].count()
driver_count_rural_city = rural_group["driver_count"].count()
```

```python
average_urban_fare = urban_group["fare"].mean()
average_suburban_fare = suburban_group["fare"].mean()
average_rural_fare = rural_group["fare"].mean()
```

```python
urban_rides_per_city = urban_group["ride_id"].count()
suburban_rides_per_city = suburban_group["ride_id"].count()
rural_rides_per_city = rural_group["ride_id"].count()
```

```python
type1 = (urban_rides_per_city,average_urban_fare)
type2 = (suburban_rides_per_city,average_suburban_fare)
type3 = (rural_rides_per_city,average_rural_fare)

data = (type1, type2, type3)
colors = ("tomato", "lightskyblue", "gold")
groups = ("Urban", "Suburban", "Rural")
drivers = (driver_count_urban_city,driver_count_suburban_city,driver_count_rural_city)

for data, color, group, driver in zip(data, colors, groups, drivers):
   x, y = data
   plt.scatter(x, y, alpha=0.8, c=color, edgecolors='black', s=driver*10, label=group)

#plt.scatter(urban_rides_per_city, average_urban_fare, s=100, c="red", facecolors="red", edgecolors="black", alpha=0.5)

plt.xlabel('Total Number of Rides (Per City)', fontsize=15)
plt.ylabel('Average Fare($)', fontsize=15)
plt.title('Pyber Ride Sharing Data (2016)',fontsize=18)

plt.grid(True)
plt.legend(loc=1)

plt.show
```

```python
# Labels for the city types of the pie chart
labels = ["Urban", "Suburban", "Rural"]

# The ride values of each section of the pie chart
ridesizes = [percent_total_rides_urban, percent_total_rides_suburban, percent_total_rides_rural]

# The ride values of each section of the pie chart
faresizes = [percent_total_fares_urban, percent_total_fares_suburban, percent_total_fares_rural]

# The ride values of each section of the pie chart
driversizes = [percent_total_drivers_urban, percent_total_drivers_suburban, percent_total_drivers_rural]

# The colors of each section of the pie chart
colors = ["tomato", "lightskyblue", "gold"]

# Separates the "Urban" section from other types
explode = (0.1, 0, 0)
```

```python
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.title('Total Fares by City Type',fontsize=18)
plt.pie(faresizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```

```python
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.title('Total Rides by City Type',fontsize=18)
plt.pie(ridesizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```

```python
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.title('Total Drivers by City Type',fontsize=18)
plt.pie(driversizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```
