```python
%matplotlib inline

# Setup Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# City and ride data from csv files
city_data_to_load = "../GitHub/Pyber_challenge/city_data.csv"
ride_data_to_load = "../GitHub/Pyber_challenge/ride_data.csv"

# Read the City and Ride Data from csv files
ride_data = pd.read_csv(ride_data_to_load)
city_data =  pd.read_csv(city_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, how='left', on='city')

# Display the data table for preview
pyber_df
```

```python
group_type_df = pyber_df.groupby(['type'])
group_type_df
```

```python
group_city_df = pyber_df.groupby(['city'])
group_city_df
```

```python
average_fare_per_type = group_type_df["fare"].mean()
average_fare_per_type
```

```python
average_fare_per_city = group_city_df["fare"].mean()
average_fare_per_city
```

```python
total_rides_per_type = group_type_df["ride_id"].count()
total_rides_per_type
```

```python
total_rides_per_city = group_city_df["ride_id"].count()
total_rides_per_city
```

```python
#plt.scatter(x, y, c=colors, s=area, alpha=0.5)
plt.scatter(total_rides_per_city, average_fare_per_city, alpha=0.5)
plt.show
```
